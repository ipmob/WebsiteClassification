```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
data = pd.read_csv('website.csv')
```

```python
data.head()
```

```python
data.category.unique()
```

```python

up_data = data.drop_duplicates(subset=['url','category'], keep='last')
up_data = up_data.reset_index(drop=True)
up_data = up_data[up_data.url != 'http://www.']
up_data = up_data.reset_index(drop=True)
```

```python
len(up_data)
```

```python
up_data.head()
```

```python
def textScrap(url):
    html = urlopen(url)
    soup = BeautifulSoup(html, "lxml")
    [s.extract() for s in soup(['style', 'script', '[document]'])]
    visible_text = soup.getText()
    return visible_text
```

```python
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
```

```python
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
```

```python
f = open('Adult.txt','a')
aaa= -1
for i in range(0,len(up_data)):
    if up_data['category'][i] == 'adult':
        try:
            url = up_data['url'][i]
            text = textScrap(url)
            tokenizer = RegexpTokenizer(r'\w+')
            tokens = tokenizer.tokenize(text)
            tokens = [x.lower() for x in tokens if not x in set(stopwords.words('english'))]
            new_items = [item for item in tokens if not item.isdigit()]
            temp = []
            for t in new_items:
                if isEnglish(t)==True:
                    temp.append(t)
            new_items = " ".join(temp)
            if aaa == -1:
                aaa = 10
            else:
                f.write('\n')
            f.write(new_items)
        except Exception:
            pass
f.close()
```

```python
from urllib.parse import urlparse

def checkUrl(url):
    try:
        p = urlparse(url)
        conn = http.client.HTTPConnection(p.netloc,timeout=10)
        conn.request('HEAD', p.path)
        resp = conn.getresponse()
        return resp.status < 400
    except:
        return False
```

```python
adult_web = []
for i in range(0,len(a)):
    if a[2][i] == 'Adult':
        if checkUrl(a[1][i]) == True:
            adult_web.append(a[1][i])
            print(i,a[1][i])
```
